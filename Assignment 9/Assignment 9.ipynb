```python
import pandas as pd
import seaborn as sns
penguins = sns.load_dataset("penguins")
penguins.head()
```

```python
 penguinsCorrelation = penguins.corr()
```

```python
penguinsCorrelation.style.background_gradient(cmap='coolwarm', axis=None).set_precision(2)
```

As you can see the bills length and depth show a negative correlation to each other. While the flipp;er length shows a bit of correlation with the bill length. Finally you can see that the flipper length shows alot of correlation with the flipperlength. Fascinating.
